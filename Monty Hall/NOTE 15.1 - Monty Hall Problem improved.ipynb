```python
from random import randrange, choice, shuffle
import matplotlib.pyplot as plt

def monty_hall_simulation(iterations, number_of_doors, change_door_strategy="random"):
    wins = 0
    wins_over_time = []

    for _ in range(iterations):
        # Generate doors
        doors = [x for x in range(number_of_doors)]
        choices = [x for x in range(number_of_doors)]

        # Generate door with prize
        k = randrange(number_of_doors)

        # Generate player guess
        l = randrange(number_of_doors)
        
#         #calculate 'empty doors'
#         doors.remove(doors[k])

        # Host shuffles the doors and opens all but 2 (excluding the prize door) to tempt the player
        if l == k:
            doors.remove(k)
            shuffle(doors)
            doors = doors[:1]
        else:
            doors = [k]
            
            
        # Player decides whether to change doors or not based on strategy
        if change_door_strategy == "random":
            change_door = choice([True, False])
        elif change_door_strategy == "always":
            change_door = True
        elif change_door_strategy == "never":
            change_door = False
        else:
            raise ValueError("Invalid change_door_strategy")

        if change_door:
            
            n = random.choice(doors)
            l = n

        if l == k:
            wins += 1
        wins_over_time.append(wins / (_ + 1))  # Track wins over time

    return wins_over_time

def main():
    iterations = 1000
    number_of_doors = 10

    # Simulate scenario where the player changes doors
    wins_change = monty_hall_simulation(iterations, number_of_doors, change_door_strategy="always")

    # Simulate scenario where the player keeps the initial choice
    wins_keep = monty_hall_simulation(iterations, number_of_doors, change_door_strategy="never")

    # Simulate scenario where the player acts erratically
    wins_erratic = monty_hall_simulation(iterations, number_of_doors, change_door_strategy="random")

    # Plot the results
    plt.plot(range(1, iterations + 1), wins_change, label="Change Doors")
    plt.plot(range(1, iterations + 1), wins_keep, label="Keep Initial Choice")
    plt.plot(range(1, iterations + 1), wins_erratic, label="Erratic Behavior")

    plt.xlabel("Number of Games Played")
    plt.ylabel("Winning Percentage")
    plt.legend()
    plt.title(f"Monty Hall Simulation ({iterations} iterations)")
    plt.show()

if __name__ == "__main__":
    main()
```
