```python
from random import randrange

def monty_hall_simulation(iterations, change_door=True):
    wins = 0
    losses = 0

    for _ in range(iterations):
        # Generate doors
        doors = [0, 1, 2]
        choices = [0, 1, 2]

        # Generate door with prize
        k = randrange(3)

        # Calculate 'empty doors'
        doors.remove(doors[k])

        # Generate player guess
        l = randrange(3)

        # Host opens an empty door to tempt the player
        if l in doors:
            doors.remove(l)

        # Player decides whether to change doors or not
        if change_door:
            choices.remove(random.choice(doors))
    
            if l in choices:
                choices.remove(l)

            n=random.choice(choices)
            l=n
            
        if l == k:
            wins += 1
        else:
            losses += 1

    return wins, losses

def main():
    iterations = 1000

    # Simulate scenario where the player changes doors
    wins_change, losses_change = monty_hall_simulation(iterations, change_door=True)

    # Simulate scenario where the player keeps the initial choice
    wins_keep, losses_keep = monty_hall_simulation(iterations, change_door=False)

    print(f"Monty Hall Simulation ({iterations} iterations):")
    print("Player changing doors:")
    print(f"Wins: {wins_change}, Losses: {losses_change}")

    print("\nPlayer keeping the initial choice:")
    print(f"Wins: {wins_keep}, Losses: {losses_keep}")

if __name__ == "__main__":
    main()
```
